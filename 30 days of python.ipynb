```python
#day 10
#loops in python
# For Loop
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print("Fruit:", fruit)
# While Loop
count = 0
while count < 5:
    print("Count:", count)
    count += 1
# Nested Loops
for i in range(1, 4):
    for j in range(1, 4):
        print("i:", i, "j:", j)
# Loop with Break and Continue
for num in range(1, 10):
    if num == 5:
        print("Breaking the loop at num =", num)
        break
    if num % 2 == 0:
        print("Skipping even number:", num)
        continue
    print("Odd number:", num)
```

```python
#day 11
#functions in python
# Defining a function
def greet(name):
    """This function greets the person with the given name."""
    print("Hello, " + name + "!")
# Calling the function
greet("Alice")
greet("Bob")
# Function with return value
def add(a, b):
    """This function returns the sum of two numbers."""
    return a + b
# Calling the function and storing the result
result = add(5, 3)
print("Sum:", result)
# Function with default parameter
def power(base, exponent=2):
    """This function returns the base raised to the exponent."""
    return base ** exponent
# Calling the function with and without the default parameter
print("Power with default exponent:", power(4))
print("Power with custom exponent:", power(4, 3))
# Function with variable-length arguments
def multiply(*args):
    """This function returns the product of all given numbers."""
    product = 1
    for num in args:
        product *= num
    return product
# Calling the function with variable-length arguments
print("Product of numbers:", multiply(2, 3, 4))


```

```python
#day 13
#lists comprehension in python
# Creating a list of squares using list comprehension
squares = [x**2 for x in range(1, 11)]
print("List of squares from 1 to 10:", squares)
# Creating a list of even numbers using list comprehension
evens = [x for x in range(1, 21) if x % 2 == 0]
print("List of even numbers from 1 to 20:", evens)
# Creating a list of uppercase strings using list comprehension
fruits = ["apple", "banana", "cherry"]
upper_fruits = [fruit.upper() for fruit in fruits]
print("List of uppercase fruits:", upper_fruits)
# Creating a list of tuples using list comprehension
tuples_list = [(x, x**2) for x in range(1, 6)]
print("List of tuples (number, square):", tuples_list)
```

```python
#day 14
#higher order functions in python
# Using map() to square each number in a list
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared Numbers:", squared_numbers)
# Using filter() to get even numbers from a list
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even Numbers:", even_numbers)
print("Even Numbers:", even_numbers)
# Using reduce() to calculate the product of all numbers in a list
from functools import reduce
product = reduce(lambda x, y: x * y, numbers)
print("Product of Numbers:", product)
# Using sorted() with a custom key to sort strings by their length
strings = ["apple", "banana", "cherry", "date"]
sorted_strings = sorted(strings, key=lambda x: len(x))
print("Strings sorted by length:", sorted_strings)
# Using any() to check if any number in the list is greater than 3
any_greater_than_3 = any(map(lambda x: x > 3, numbers))
print("Any number greater than 3:", any_greater_than_3)
```

```python
#day 21
#classes and objects in python
# Defining a class
class Dog:
    # Constructor method to initialize attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # Method to make the dog bark
    def bark(self):
        return f"{self.name} says Woof!"
    # Method to get the dog's age
    def get_age(self):
        return f"{self.name} is {self.age} years old."
# Creating an object (instance) of the Dog class
my_dog = Dog("Buddy", 3)
# Accessing attributes and methods of the object
print(my_dog.bark())
print(my_dog.get_age())
# Creating another object of the Dog class
another_dog = Dog("Max", 5)
print(another_dog.bark())
print(another_dog.get_age())
```

```python
%pip install fastapi uvicorn pydantic

#day28
#python API
from fastapi import FastAPI
from pydantic import BaseModel
from pymongo import MongoClient
from bson import ObjectId

app = FastAPI()

client = MongoClient("mongodb://localhost:27017/")
db = client["simple_api"]
col = db["people"]

class Person(BaseModel):
    name: str
    age: int

def serialize(doc):
    return {"id": str(doc["_id"]), "name": doc["name"], "age": doc["age"]}

@app.post("/add")
def add_person(person: Person):
    res = col.insert_one(person.dict())
    return serialize(col.find_one({"_id": res.inserted_id}))

@app.get("/all")
def all_people():
    return [serialize(i) for i in col.find()]


```
